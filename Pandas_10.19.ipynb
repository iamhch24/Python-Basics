### 3. 각 영화별 평균 평점을 movieid, title, 평균평점 순으로 출력하세요

```python
movie_mean= df_rating.groupby("movieId")["rating"].mean()
movie_mean# 시리즈
result_df = movie_mean.to_frame() # 시리즈를 DataFrame으로 만들어줌
result_df.columns = ['average'] # 컬럼을 설정

df = pd.merge(df_m, result_df, left_on="movieId", right_index = True, how="inner")
df[["movieId","title","average"]]
```

### 4. 평균평점이 가장 높은 영화의 제목을 출력하세요 

```python
df.loc[ df["average"] == df["average"].max(),'title']     
```

### 5. Comedy 영화 중 가장 평점이 낮은 영화 제목 출력 (동물이 있는 경우 모두 출력하세요)

```python
comedyDf = df.loc[df["genres"].str.contains("Comedy")]
comedyDf.loc[comedyDf["average"]==comedyDf["average"].min(),'title']
```

### 6. 2015년도에 평가된 모든 Romance 영화의 평균 평점

```python
import time

df2 = pd.merge(df_m,df_rating,left_on="movieId",right_on="movieId",how="inner")
result = df2.loc[df2["genres"].str.contains("Romance")]
result
result_df = pd.DataFrame(columns=["평점"])
result_df
start = time.mktime(time.strptime("2015-01-01 00:00:00","%Y-%m-%d %H:%M:%S"))
end = time.mktime(time.strptime("2016-01-01 00:00:00","%Y-%m-%d %H:%M:%S"))
print(result.loc[ (result["timestamp"] >=start) & (result["timestamp"] < end), "rating"].mean() )
```

### ※ 2.7 Pandas Groupby

```python
data = [
    ['Riders',1,2014,876],
    ['Riders',2,2015,789],
    ['Devils',2,2014,863],
    ['Devils',3,2015,673],
    ['Kings',3,2014,741],
    ['Kings',4,2015,812],
    ['Kings',1,2016,756],
    ['Kings',1,2017,788],    
    ['Riders',2,2016,694],
    ['Royals',4,2014,701],
    ['Royals',1,2015,804],
    ['Riders',2,2017,690] ]
columns =['Team','Rank','Year','Points']

df = pd.DataFrame(data = data, index = None, columns = columns)
df
```

```python
df.groupby("Year")["Points"].sum()
```

```python
df.groupby("Team")["Rank"].mean()
```

```python
h_index = df.groupby(["Team","Year"])["Points"].sum()
h_index
```

```python
h_index.unstack()
```

```python
h_index.swaplevel()
```

```python
Teams = df.groupby('Team')
for name in Teams:
    print(name)
```

```python
Teams.get_group("Devils")
```

```python
Teams.agg(sum)
```
