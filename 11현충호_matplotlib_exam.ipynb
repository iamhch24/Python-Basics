```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline


x = np.linspace(0.1,16,100)
x2 = np.linspace(0.1,11,40)

fig=plt.figure()
ax = fig.add_axes([0,0,1,4])
ax.plot(x, np.log2(x), label='y=log(x)', color = 'orange')
ax.plot(x, x, label='y=x', color = 'black')
ax.plot(x, x*np.log2(x), label='y=xlog(x)', color = 'yellow')
ax.plot(x, x**2, label='y=x**2', color='green')
ax.plot(x,x**3, label='y=x**3', color='blue')
ax.plot(x,2**x, label='y=2**x', color='red')
ax.legend()

ax2 = fig.add_axes([0.1,2.5,0.7,1])
ax2.plot(x2, np.log2(x2), label='y=log(x)', color = 'orange')
ax2.plot(x2, x2, label='y=x', color = 'black')
ax2.plot(x2, x2*np.log2(x2), label='y=xlog(x)', color = 'yellow')
ax2.plot(x2, x2**2, label='y=x**2', color='green')
ax2.plot(x2,x2**3, label='y=x**3', color='blue')
ax2.plot(x2,2**x2, label='y=2**x', color='red')
ax2.legend()
```
