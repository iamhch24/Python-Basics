```python
import random

mean = 70
std = 10
l = []
for i in range(10):    
    if random.randint(0, 1) == 0:
        score = mean + int(std * random.random())
    else:
        score = mean - int(std * random.random())
    #print(score)
    l.append(score)

import matplotlib.pyplot as plt
#matplotlib 패키지 한글 폰트 설정 시작
#plt.rc('font', family='Malgun Gothic') #윈도우, 구글 콜랩
mpl.rc('font', family='Malgun Gothic') #한글 폰트 설정
#matplotlib 패키지 한글 폰트 설정 끝
plt.scatter(range(10), l)
plt.title('분포도')
plt.xlabel('아이디')
plt.ylabel('점수')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline


x = np.linspace(0.1,16,100)
x2 = np.linspace(0.1,4,20)

fig=plt.figure()
ax = fig.add_axes([0,0,1,4])
ax.plot(x, np.log(x), label='y=log(x)', color = 'orange')
ax.plot(x, x, label='y=x', color = 'black')
ax.plot(x, x*np.log(x), label='y=xlog(x)', color = 'yellow')
ax.plot(x, x**2, label='y=x**2', color='green')
ax.plot(x,x**3, label='y=x**3', color='blue')
ax.plot(x,2**x, label='y=2**x', color='red')
ax.legend()

ax2 = fig.add_axes([0.2,2.5,0.6,1])
ax2.plot(x2, np.log(x2), label='y=log(x)', color = 'orange')
ax2.plot(x2, x2, label='y=x', color = 'black')
ax2.plot(x2, x2*np.log(x2), label='y=xlog(x)', color = 'yellow')
ax2.plot(x2, x2**2, label='y=x**2', color='green')
ax2.plot(x2,x2**3, label='y=x**3', color='blue')
ax2.plot(x2,2**x2, label='y=2**x', color='red')
ax2.legend()
```
