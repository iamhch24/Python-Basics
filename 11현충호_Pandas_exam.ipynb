'''

'''



```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

#### 문제1)  남자(male) 이면서 Pclass 가 1인 승객은 몇 명인가요?

```python
df.loc[(df['Sex'] == 'male') & (df['Pclass']==1)]['PassengerId'].count()
```

####  문제2) 남자이면서 Pclass 가 1이고 Fare가 50 이상인 승객은 몇 명인가요?

```python
df.loc[(df['Sex'] == 'male') & (df['Pclass']==1) & (df['Fare']>=50) ]['PassengerId'].count()
```

#### 문제3) 새로운 열(sex_code)을 만들어서 남자인 경우는 코드 0, 여자인 경우는 1로 저장

```python
df['sex_code'] = df.Sex.map({'male':0, 'female':1})
df.head()
```

#### 문제4) 성별 생존자 수의 합(남자 생존자 수의 합, 여자 생존자수의 합) - 생존자는 Survived 의 값이 1인 것.

```python
df.loc[df['Survived']==1].groupby('Sex')['Survived'].sum()
```
